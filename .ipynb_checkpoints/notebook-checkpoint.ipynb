## Dataset overview
The dataset chosen for this Machine Learning project is a Wine Reviews dataset from Kaggle.com (https://www.kaggle.com/zynicide/wine-reviews/version/4). The dataset includes almost 130000 professional reviews about different types of wines. It contains 10 features that includes a score in a scale from 1 to 100 rated by the members of WineEnthusiast magazine, title of the wine, variety - what grapes were used for brewing the wine, description which includes a quick summary describing the taste and smell of the wine, location information that includes country, region and province and the name of winery, designation describes a vineyard within the winery. Finally, it includes the price of the bottle and the reviewer details such as name and twitter handle.

## Problem
The main objective of this data analysis is to automatically predict wine ranking based on the other features available - description, price, type of wine, etc. To do so, we will train a regression model based on the features available in the dataset. The problem is challenging since the dataset mainly includes textual dataa and only two numerical features are score and price.

## Dataset cleaning
After briefly looking into dataset we should first figure out whether we do have empty or missing data. Some features as reviewer name have to be removed for anonimization purposes and there are features that do not carry any meaning useful for predictions. Description could be a good predictor if most common keywords are present in highest rated wines, but we would need to transform the description by removing stopwords, most common words, stemming, and do feature extraction to make it useful. We assume the description contents are important and some specific words can influence the ranking score.

Since majority of the features are text-based, these will need to be one-hot encoded before analyzing. To extract features from the review description, TF-IDF vectorizer will be used to remove stop words and count TF-IDF scores for each word.

#### First of all, check if there are any missing data

```python
data.isnull().any()
```

We are going to drop the columns that I assume will be not useful for analysis and then remove rows that have missing data. We will make a copy of the original dataset to preserve original dataset and avoid making mistakes by running the selection twice.

```python
wine_reviews = data.copy();

# drop some useless columns
wine_reviews.drop(data.columns[0:1], axis=1, inplace=True) # drop Unnamed: 0 column which is meaningless and contains incremental id
wine_reviews.drop("taster_name", axis=1, inplace=True) # assuming reviewer name is not a strong predictor
wine_reviews.drop("taster_twitter_handle", axis=1, inplace=True) # same as above, twitter handle is not imporant

wine_reviews.drop("winery", axis=1, inplace=True) # too many wineries in the world to be able hot-1-encode
wine_reviews.drop("designation", axis=1, inplace=True) # deisgnation is a vineyard in winery, so it is not important
wine_reviews.drop("region_1", axis=1, inplace=True) # lot's of empty variables, just drop the column
wine_reviews.drop("region_2", axis=1, inplace=True) # lot's of empty variables, just drop the column

wine_reviews.drop("title", axis=1, inplace=True) # title is unique for all so it is not a good predictor

# now drop rows that have n/a
wine_reviews = wine_reviews.dropna(subset=["country", "variety", "price", "province"])

# At this point we should not have any empty data
wine_reviews.isnull().any()
```

To reduce the dataset and speedup training, we will only select top 5 countries and top 30 provinces which will leave us with 93303 rows out of 130000 what we had originally.

```python
wine_reviews = wine_reviews.loc[data['country'].isin(pd.value_counts(data.country).head(5).keys())] # select top 5 countries
wine_reviews = wine_reviews.loc[data['province'].isin(pd.value_counts(data.province).head(30).keys())] # select top 30 province
                 
wine_reviews.shape
```

```python
wine_reviews.dtypes
```

### Data Transformation and  combination of the  features

Since the dataset mostly contain text features (except price and points, see above), we need to transform the features before we are able to plug into a machine learning regressor to estimate the point score of the bottle of the wine. 

First of all, we need to transform review descriptions using TF-IDF transformer to index most common words and its counts. This will allow to analyze description in a machine learning model and predict our result.

For the remaining text features we need to do a hot-one-encoding. Since we will endup with a lot of new dimensions and high cardinality, we should do dimensionality reduction using Principal component analysis (PCA) to reduce dimensionality. 

Finally we will combine both feature sets using Feature Union so we could apply machine learning model on the two differently transformed sets.


```python
# Organize our data into X (features) and Y (price, which we will be trying to predict)
X = wine_reviews.drop("points", axis=1) # get all except the price
Y = wine_reviews.points # get just the price

### Extract features from description and apply TF-IDF
class DescriptionExtractor(BaseEstimator, TransformerMixin):

    def transform(self, X, y=None):
        vectorizer = TfidfVectorizer(stop_words='english')
        return vectorizer.fit_transform(X['description'], y)
        
    def fit(self, X, y=None):
        return self

### Hot - one encoding for all other columns and do a principal component analysis
class TextColumnEncoder(BaseEstimator, TransformerMixin):
    
    def transform(self, X, y=None):
        # do one hot encoding without description
        one_hot_encoded_X = pd.get_dummies(X.drop("description", axis=1), columns=["country", "variety", "province"], sparse=True)
        pca = PCA(n_components=2)
        # Now we have a lot of dimensions, we need to do a dimensionality reduction with PCA
        return pca.fit_transform(one_hot_encoded_X, y)

    def fit(self, X, y=None):
        return self
    
# Combine PCA features with TF-IDF vectorizer
combined_features = FeatureUnion(transformer_list=[
            ('description', DescriptionExtractor()),
            ('other_features', TextColumnEncoder())
        ], transformer_weights={
            'description': 0.4,
            'other_features': 0.6,
        },)


X_combined_features = combined_features.fit_transform(X, Y)
```


### Machine learning models that have been selected to evaluate:
* Linear Regression
* Decision Trees
* Gradient Boosting
* Random Forest
* Stochastic Gradient Descent (SGD)

* K Nearest Neighbors
* Support Vector Machines (SVM)

### After further testing, training and review I have decided to remove several models
* K Nearest Neighbors - KNN suffers from high dimentional data as it uses Euclidean distance as a distance metric which becomes slow and meaningless when the dimension of the data increase significantly.
* Support Vector Machines - Same as K Nearest Neighbors, SVM also suffers from high dimensionality.
* Stochastic Gradient Descent - The model performed very poor, and predictions were very far off, therefore it was excluded from the futher analysis

### Method to batch evaluate multiple machine learning models and generate charts

```python

# Train dataset using all algorithns and plot into the chart
def trainDatasetAndEvaluate(X_train, X_test, y_train, y_test):
    scores = []

    # Load up models
    models = []
    models.append(('Linear', LinearRegression()))
    models.append(('Decision Tree', DecisionTreeRegressor()))
    models.append(('Gradient Boosting (ensemble)', GradientBoostingRegressor()))
    models.append(('Random Forest (ensemble)', RandomForestRegressor(n_jobs = -1, n_estimators=10, min_samples_split=5)))
    #models.append(('SGD',SGDRegressor(loss="huber", average=True, max_iter=1000, tol=None)))
    #models.append(('SVM ', SVR()))
    #models.append(('K Neighbors', KNeighborsRegressor(n_neighbors = 3)))

    fig = plt.figure(1, figsize=(10, 10))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    # loop all models
    index = 0           
    for name, regr in models:
        # Train our classifier
        model = regr.fit(X_train, y_train)
        # Make predictions
        predicted = regr.predict(X_test)
        
        # Evaluate each model
        r2 = metrics.r2_score(y_test, predicted)
        rmse = np.sqrt(metrics.mean_squared_error(y_test, predicted))
        mae  = metrics.mean_absolute_error(y_test, predicted)            
        variance = metrics.explained_variance_score(y_test, predicted)
                       
        scores.append((name, r2, rmse, mae, variance))

        # Print metrics
        print("%s R2 score %.2f" % (name, r2))
        print("Root Mean squared error: %.2f" % rmse)
        print("Mean absolute error: %.2f" % mae)
        print("Explained variance score: %.2f" % variance)

        ax = fig.add_subplot(2,2, index + 1)
        
        # Plot outputs
        ax.scatter(y_test, predicted, color='blue')
        plt.title(name)
        plt.ylabel('Predicted')
        plt.xlabel('Actual')
        plt.xticks(())
        plt.yticks(())

        print("\n")
        index = index + 1
    

    plt.show()
    
    # Plot comparisons 
    fig = plt.figure(2, figsize=(10, 10))
    fig.subplots_adjust(hspace=1.4, wspace=0.4)

    # generate a bar chart with scores
    # sort in-place from highest to lowest
    scores.sort(key=lambda tupl: tupl[1], reverse=True) 
    names = list(zip(*scores))[0]
    score = list(zip(*scores))[1]

    ax = fig.add_subplot(2,2,1)
    ax.bar(np.arange(len(scores)), score, align='center', alpha=0.5)
    plt.xticks(np.arange(len(scores)), names, rotation=90)
    plt.title('R2 score')
    plt.ylabel('Score')
    plt.xlabel('Model')

    scores.sort(key=lambda tupl: tupl[2], reverse=True) 
    names = list(zip(*scores))[0]
    score = list(zip(*scores))[2]

    ax = fig.add_subplot(2,2,2)
    ax.bar(np.arange(len(scores)), score, align='center', alpha=0.5)
    plt.xticks(np.arange(len(scores)), names, rotation=90)
    plt.title('Root Mean squared error')
    plt.ylabel('Score')
    plt.xlabel('Model')

    scores.sort(key=lambda tupl: tupl[3], reverse=True) 
    names = list(zip(*scores))[0]
    score = list(zip(*scores))[3]

    ax = fig.add_subplot(2,2,3)
    ax.bar(np.arange(len(scores)), score, align='center', alpha=0.5)
    plt.xticks(np.arange(len(scores)), names, rotation=90)
    plt.title('Mean absolute error')
    plt.ylabel('Score')
    plt.xlabel('Model')

    scores.sort(key=lambda tupl: tupl[4], reverse=True) 
    names = list(zip(*scores))[0]
    score = list(zip(*scores))[4]

    ax = fig.add_subplot(2,2,4)
    ax.bar(np.arange(len(scores)), score, align='center', alpha=0.5)
    plt.xticks(np.arange(len(scores)), names, rotation=90)
    plt.title('Variance')
    plt.ylabel('Score')
    plt.xlabel('Model')

    plt.show()
```

## Final Step: Regressor training and evaluation


```python
# now split the dataset to training and test set by 30%:70% split
X_train, X_test, y_train, y_test = train_test_split(X_combined_features, Y, test_size=0.3)

trainDatasetAndEvaluate(X_train, X_test, y_train, y_test);
```

## Results
Selected regression models have been trained based on the wine reviews dataset. Looking into the scatter charts above we can see that each of the selected machine learning algorithms performed quite well while trying to predict wine score based on the features such as country, winery, variety, price and province. The graph produced is uniform. K Nearest Neighbors, Support Vector Machines were excluded from the charts to avoid misrepresentation of the results as these either scored very poorly or failed to complete the training. 

To evaluate different regression models several metrics were used that are suitable for regression algorithms:
* Root Mean Squared Error
* Mean Absolute Error 
* R2 Score

The model which generated the highest R2 score of 0.62 (maximum score is 1.0) was Random Forrests. Random Forests are among most powerful machine learning algorithms available within scikit-learn so it is not unexpected that it performed best. However, it took way longer to train the model compared with the other models. Linear regression and Gradient Boosting scored very similarly by having 0.56 and 0.52 R2 scores respectively. Finally, Descicion Tree performed worst by having just 0.34 R2 score.

Other evaluation metrics such as root mean squared error and mean absolute error displayed similar outcomes as the R2 score. Root mean squared error is 2.47, 2.06, 2.01 and 1.97 for Decision Trees, Linear regression, Gradient Boosting and Random Forrests respectively (lower is better). Finally, Mean absolute error is 1.78, 1.55, 1.59, 1.42 for Decision Trees, Linear regression, Gradient Boosting and Random Forrests respectively (lower is better).


## Summary
Taking everything into consideration, Wine reviews dataset revealed that it is possible to predict the score (in a range 1-100) based on wine description and wine features (country, province, type, variety) quite accurately. Before running through the machine learning algorithm the dataset had to be cleaned by removing several features and removing rows with missing data. Text features (country, province, type, variety) were encoded using Hot-One and to reduce dimensionality Principal Component Analysis (PCA) was applied. In addition, review descriptions were vectorized with TF-IDF vectorizer to extract meaning from the review description. Finally, text features (country, province, type, variety), numeric feature (price) and review descriptions were combined into a single dataset using FeatureUnion.

The dataset was trained on several different machine learning models and Random Forrest regressor has predicted most accurate results.

On the other hand, it was interesting to see whether using the same features, price instead of the score could be predicted by switching Y label in calculation. Unfortunately this gave us not so good results making it more difficult to predict the price compared to the score. This could mean that some wine prices may be too expensive or too cheap compared with the quality and taste of the wine.
